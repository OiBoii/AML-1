```python
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import warnings
warnings.filterwarnings("ignore")
```

(a)

```python
img = Image.open('monalisa.jpg').convert('RGB')
```

```python
arr = np.asarray(img)
```

```python
plt.figure(figsize = (12,9))
plt.imshow(arr)
plt.show()
```

Random 5000 sample coordinates:

(b)

```python
height, width, k  = arr.shape
```

```python
n = 5000
train_x = np.random.uniform(0,width-1,n).round()
train_y = np.random.uniform(0,height-1,n).round()
train_xy = np.array(list(zip(train_x, train_y)))
```

(c)

```python
def pxl(tp, img):
    x = int(tp[0])
    y = int(tp[1])
    return img[y][x]
```

```python
train_img = np.array([[[255, 255, 255] for i in range(width)] for j in range(height)], dtype="uint8")
```

```python
train_img.shape
```

```python
for px in train_xy:
    x = int(px[0])
    y = int(px[1])
    train_img[y][x] = arr[y][x]
```

```python
plt.figure(figsize = (12,9))
plt.imshow(train_img)
plt.show()
```

(d)

```python
train_label = np.array(list(map(lambda x: list(pxl(x, arr)), train_xy)))
```

```python
from sklearn.ensemble import RandomForestRegressor as RFR
```

```python
regr = RFR(max_depth=20, random_state = 0, n_estimators=30)
regr.fit(train_xy, train_label)
```

```python
test_xy = np.array([[x, y] for y in range(height) for x in range(width)])
```

```python
prediction = regr.predict(test_xy)
prediction = prediction.round()
```

```python
pred_m = prediction.reshape(height, width, 3)
```

```python
pred_m = np.array(pred_m,dtype="uint8")
```

```python
plt.figure(figsize = (12,9))
plt.imshow(pred_m)
plt.show()
```

(e)

    (i)

```python
plt.figure(figsize = (12,9))

pos = 1
for i in [1,2,3,5,10,15]:
    regr = RFR(max_depth=i, random_state = 0,n_estimators=1)
    regr.fit(train_xy, train_label)
    test_xy = np.array([[x, y] for y in range(height) for x in range(width)])
    prediction = regr.predict(test_xy)
    prediction = prediction.round()
    pred_m = prediction.reshape(height, width, 3)
    pred_m = np.array(pred_m, dtype="uint8")
    plt.subplot(230+pos)
    plt.imshow(pred_m)
    plt.title("Depth = " + str(i))
    pos += 1
plt.show()
```

* As the depth increasing, the approximate of image becomes better and better
* The reason is that, as depth increasing, the regressor will use more blocks to approximate the image so the image becomes more clear. 

    (ii)

```python
plt.figure(figsize = (12,9))

pos = 1
for j in [1,3,5,10,10,100]:
    regr = RFR(max_depth=7, random_state = 0,n_estimators=j)
    regr.fit(train_xy, train_label)
    test_xy = np.array([[x, y] for y in range(height) for x in range(width)])
    prediction = regr.predict(test_xy)
    prediction = prediction.round()
    pred_m = prediction.reshape(height, width, 3)
    pred_m = np.array(pred_m, dtype="uint8")
    plt.subplot(230+pos)
    plt.imshow(pred_m)
    plt.title("Num of Trees = " + str(j))
    pos += 1
plt.show()
```

* As the number of trees getting larger, the image becomes smoother. 
* The reason is that the regressor will have less bias when more trees are in the random forest.

    (iv)

```python
plt.figure(figsize = (12,9))

pos = 1
for j in [5,10,25,100,250]:
    regr = RFR(max_depth=10, random_state = 0,
               n_estimators=100, max_leaf_nodes=j)
    regr.fit(train_xy, train_label)
    test_xy = np.array([[x, y] for y in range(height) for x in range(width)])
    prediction = regr.predict(test_xy)
    prediction = prediction.round()
    pred_m = prediction.reshape(height, width, 3)
    pred_m = np.array(pred_m, dtype="uint8")
    plt.subplot(230+pos)
    plt.imshow(pred_m)
    plt.title("Max_leaf_nodes = " + str(j))
    pos += 1
plt.show()
```

```python
plt.figure(figsize = (12,9))

pos = 1
for j in [5,10,25,50,100]:
    regr = RFR(max_depth=10, random_state = 0,
               n_estimators=100, min_samples_leaf=j)
    regr.fit(train_xy, train_label)
    test_xy = np.array([[x, y] for y in range(height) for x in range(width)])
    prediction = regr.predict(test_xy)
    prediction = prediction.round()
    pred_m = prediction.reshape(height, width, 3)
    pred_m = np.array(pred_m, dtype="uint8")
    plt.subplot(230+pos)
    plt.imshow(pred_m)
    plt.title("min_samples_leaf = " + str(j))
    pos += 1
plt.show()
```

(f)

    (i)

For an example, at a certain root, we picked x as the feature at this split point. <br>
Define variables: <br>
T: The threshold. <br>
pt_left: if the sample's x coordinate > T, it is in pt_left. <br>
pt_right: if the sample's x coordinate <= T, it is in pt_left. <br>
left_mean: mean of pt_left. <br>
right_mean: mean of pt_right. <br>

To obtain the threshold. We want to minimize the sum of the error of |pt_left - left_mean|^2 and |pt_right - right_mean|^2 

    (ii)

* The image are looks many rectangles with different colors because the decision tree can only use one feature to split each time. At the boundary between two rectangles, color looks like a mixture of the color of two rectangles. The reason is that random forest will take the mean of the outputs from all the trees. 

    (iii)

* Number of patches could be 2^max_depth at the most. 

    (iv)

* If there are n trees, the total leafs will be n*2^max_depth. The color of patch will be the average of n leafs, so the total number of patches can be n*2^max_depth choose n.
